### Image Blurring

```python
#  to remove the noice we can use blur
# blur , gaussion blur, median blur, bilateral filter -> different functions of opencv
#  replacing the pixel value with the average of the neighboring pixel values ->HIGH LEVEL INTIUITION

import os
import cv2
img =cv2.imread(os.path.join('.','data','freelancer.jpg'))
ksize=7 #larger the kernel size more blurrier the image
img_blur =cv2.blur(img, (ksize,ksize) )
img_blur_gaussian=cv2.GaussianBlur(img,(ksize,ksize),0)
img_blur_median=cv2.medianBlur(img,ksize)

cv2.imshow('original',img)
cv2.imshow('Blurred',img_blur)
cv2.imshow('Gaussian Blurred',img_blur_gaussian)
cv2.imshow('Median Blurred',img_blur_median)
cv2.waitKey(0)

```

### Threshold

convert the image into a binary image (most common isecase of threseholding)
different types of image exist 

to creat3e semantic segmentation we use threhsold, so separate the objects , for object detection

Improve it using blurs and different cv2 functions

sometimes global threshold is not good enough becuse 1 global threshold is not enough to separate the objects or handwriting , so we use adaptive threshold

**Global threshold**

```python
import os
import cv2

img= cv2.imread(os.path.join('.','data','bear.jpg'))

img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# 80  is global threshold value, if pixel value is greater than 80 then it will be set to 255 otherwise it will be set to 0
ret,thresh=cv2.threshold(img_gray,80,255,cv2.THRESH_BINARY)

thresh=cv2.blur(thresh,(10,10),0)

ret,thresh=cv2.threshold(thresh,80,255,cv2.THRESH_BINARY)

cv2.imshow('original',img)
cv2.imshow('threshold',thresh)
cv2.waitKey(0)
```

**Adaptive threshold**

```python
import os
import cv2

img=cv2.imread(os.path.join('.','data','h2.jpg'))
img= cv2.resize(img,(384,512))

img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# ret,img_thresh=cv2.threshold(img,60,255,cv2.THRESH_BINARY)

#  2 ways for adaptive thresholding - mean and gaussian

# we can specify the block size and the constant value to be subtracted from the mean or gaussian value

# for each pixel, the threshold value is calculated based on the mean or gaussian of the neighboring pixel values

thresh=cv2.adaptiveThreshold(img_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,21,30)
# 21 and 30 are the block size and constant value respectively, we can experiment with different values to see the effect on the thresholded image


cv2.imshow('original',thresh)
cv2.waitKey(0)
```

### Edge Detector

There are 3 different types of edge detectors laplace , canny edge detector , sobel derivatives

```python
import os 
import cv2
import numpy as np
img= cv2.imread(os.path.join('.','data','basketball.jpg'))
img= cv2.resize(img,(400,512))


img_edge= cv2.Canny(img,100,200) # 100 and 200 are the lower and upper threshold values for edge detection, we can experiment with different values to see the effect on the edge detected image

# hysterisis thresholding is used in canny edge detection to connect the edges detected by the lower and upper threshold values, it helps to reduce the noise and improve the edge detection results

img_edge_d=cv2.dilate(img_edge,np.ones((3,3),dtype=np.uint8))
# dilation is a morphological operation that is used to increase the size of the edges detected by the canny edge detection, it helps to connect the edges and make them more visible, we can experiment with different kernel sizes to see the effect on the dilated edge image

img_edge_e=cv2.erode(img_edge_d,np.ones((3,3),dtype=np.uint8),iterations=1)
# erosion is a morphological operation that is used to decrease the size of the edges detected by the canny edge detection, it helps to remove the noise and make the edges more thin, we can experiment with different kernel sizes to see the effect on the eroded edge image

# cv2.imshow('original',img)
cv2.imshow('edge',img_edge)
cv2.imshow('edge_dilated',img_edge_d)
cv2.imshow('edge_eroded',img_edge_e)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Drawing Shapes

```python
import os
import cv2

img=cv2.imread(os.path.join('.','data','whiteboard.jpg'))

# line
cv2.line(img,(90,120),(150,250),(0,255,0),3) # (0,255,0) is the color of the line in BGR format and 5 is the thickness of the line
print(img.shape)
# Rectangle
cv2.rectangle(img,(200,100),(300,200),(255,0,0),2) # (255,0,0) is the color of the rectangle in BGR format and 5 is the thickness of the rectangle
# by -ve thickness we can fill the rectangle with color
# Circle

cv2.circle(img,(300,250),15,(0,0,255),3) # (0,0,255) is the color of the circle in BGR format and 5 is the thickness of the circle , 15 is the radius of the circle
cv2.putText(img,"hey !",(500,250),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2) # (255,255,0) is the color of the text in BGR format and 2 is the thickness of the text ,1 is the font scale of the text
cv2.imshow('original',img)
cv2.waitKey(0)
```
